### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, "html5lib")
c = []
news = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find_all("em")[-1].text, d.find_all('a')[-1].text)
    c.append(d.find_all("em")[-1].text)
    news.append(str(d.find_all('a')[-1].text).replace(u'\u3000', u' '))
```

```python
len(c), len(news)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# 用dictionary
classified = {}
for i in range(len(c)):
    if c[i] in classified:
        classified[c[i]].append(news[i])
    else:
        classified[c[i]] = []
        classified[c[i]].append(news[i])
classified
```

```python
classified.keys()
```

③ 哪一個類別的文章最多

```python
class_len = []
for key in classified:
    class_len.append(len(classified[key]))
    
list(classified.keys())[class_len.index(max(class_len))]
```
